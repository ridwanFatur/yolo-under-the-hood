```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def pre_transform(im: list[np.ndarray], size=(640, 640)) -> list[np.ndarray]:
    new_h, new_w = size
    output = []

    for img in im:
        h, w = img.shape[:2]

        scale = min(new_w / w, new_h / h)

        resized = cv2.resize(
            img,
            (int(round(w * scale)), int(round(h * scale))),
            interpolation=cv2.INTER_LINEAR,
        )

        output.append(resized)
    return output
```

```python
image_path = "bus.jpg"

img = cv2.imread(image_path)
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
transformed = pre_transform([img])
```
